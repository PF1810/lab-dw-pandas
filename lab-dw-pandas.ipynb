### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

```python
#Compute summary statistics such as mean, median, mode, standard deviation, and quartiles to understand the central tendency
# and distribution of the data for numerical columns

import pandas as pd
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
insurance_df = pd.read_csv(url)
print(insurance_df.head())



```

```python
numerical_columns = ["Customer Lifetime Value", "Income", "Monthly Premium Auto", "Number of Open Complaints", "Total Claim Amount"]
for column in numerical_columns:
    insurance_df[column] = pd.to_numeric(insurance_df[column], errors="coerce")
```

```python
summary_stats = insurance_df.describe().T
print(summary_stats)
```

```python
mean_values = insurance_df[numerical_columns].mean()
print("Mean values:")
print(mean_values)
```

```python
median_values = insurance_df[numerical_columns].median()
print("Median values:")
print(median_values)
```

```python
mode_values = insurance_df[numerical_columns].mode().iloc[0]
print("Mode values:")
print(mode_values)
```

```python
std_dev_values = insurance_df[numerical_columns].std()
print("Standard deviation values:")
print(std_dev_values)
```

```python
q1_values = insurance_df[numerical_columns].quantile(0.25)
q3_values = insurance_df[numerical_columns].quantile(0.75)
print("25th Percentile (Q1) values:")
print(q1_values)
print("75th Percentile (Q3) values:")
print(q3_values)
```

```python
# Combine all summary statistics 
summary_stats['mean'] = mean_values
summary_stats['median'] = median_values
summary_stats['mode'] = mode_values
summary_stats['std_dev'] = std_dev_values
summary_stats['q1'] = q1_values
summary_stats['q3'] = q3_values

# Rename columns for clarity
summary_stats.rename(columns={'25%': 'q1', '50%': 'median', '75%': 'q3'}, inplace=True)

# Selecting only numerical columns for the final statistics table
summary_stats = summary_stats.loc[numerical_columns]

# Reorder and rename columns for clarity
summary_stats = summary_stats[['mean', 'median', 'mode', 'std_dev', 'q1', 'median', 'q3']]

print(summary_stats)

```

### Exercise 1

The marketing team wants to know the top 5 less common customer locations. Create a pandas Series object that contains the customer locations and their frequencies, and then retrieve the top 5 less common locations in ascending order.

```python
# Your code here
import pandas as pd

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
insurance_df = pd.read_csv(url)
insurance_df

```

```python
# Count the frequency of each customer location (state)
location_counts = insurance_df['ST'].value_counts()

print(location_counts)
```

```python
# Group by 'ST' and calculate the size of each group
location_counts = insurance_df.groupby('ST').size()

print("Customer locations and their frequencies:")
print(location_counts)
```

```python
# Get the top 5 least common locations
least_common_locations = location_counts.nsmallest(5)

print(least_common_locations)
print("Top 5 least common customer locations in ascending order:")
print(least_common_locations)
```

```python
sorted_location_counts = location_counts.sort_values()

least_common_locations = sorted_location_counts.head(5)

print("Top 5 least common customer locations in ascending order:")
print(least_common_locations)
```

### Exercise 3

The sales team wants to know if customers with Personal Auto have a lower income than those with Corporate Auto. How does the average income compare between the two policy types?

- Use *loc* to create two dataframes: one containing only Personal Auto policies and one containing only Corporate Auto policies.
- Calculate the average income for each policy.
- Print the results.

```python
# Your code here
import pandas as pd

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
insurance_df = pd.read_csv(url)

insurance_df
```

```python
# Create a DataFrame for Personal Auto policies
personal_auto_df = insurance_df.loc[insurance_df['Policy Type'] == 'Personal Auto']

# Create a DataFrame for Corporate Auto policies
corporate_auto_df = insurance_df.loc[insurance_df['Policy Type'] == 'Corporate Auto']
```

```python
# Calculate the average income for Personal Auto policies
average_income_personal_auto = personal_auto_df['Income'].mean()

# Calculate the average income for Corporate Auto policies
average_income_corporate_auto = corporate_auto_df['Income'].mean()
```

```python
# Print the average incomes for comparison
print(f"The average income for Personal Auto policies is: ${average_income_personal_auto:.2f}")
print(f"The average income for Corporate Auto policies is: ${average_income_corporate_auto:.2f}")
```

Similarly, the 25th percentile represents the value below which 25% of the data falls. When we talk about the top 25%, we are referring to the values that fall above the 75th percentile, which represent the top quarter of the data. On the other hand, when we talk about the bottom 25%, we are referring to the values that fall below the 25th percentile, which represent the bottom quarter of the data.
